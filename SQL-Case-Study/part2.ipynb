```python
import sqlite3
from sqlite3 import Error
import pandas as pd
```

```python
def connect(file):
    db = None
    try:
        db = sqlite3.connect(file)
    except Error as e:
        print(e)
    return db

 
def displayQuery(db, query):
    cur = db.cursor()
    cur.execute(query)
    for row in cur.fetchall():
        print(row)
```

```python
db = connect("sqlite_db_pythonsqlite.db")
```

```python
q10 = """
SELECT f.name, (mem_uses.memct * membercost + guest_uses.guestct * guestcost) AS total_revenue 
FROM Facilities AS f 
LEFT JOIN (SELECT facid, COUNT(*) AS memct FROM Bookings WHERE memid != 0 GROUP BY facid) AS mem_uses ON mem_uses.facid=f.facid 
LEFT JOIN (SELECT facid, COUNT(*) as guestct FROM Bookings WHERE memid = 0 GROUP BY facid) AS guest_uses ON guest_uses.facid=f.facid
GROUP BY f.name
HAVING total_revenue < 1000
ORDER BY total_revenue
"""
displayQuery(db,q10)
```

```python
q11 = """
SELECT (m1.surname ||', '|| m1.firstname) AS fullname,  (m2.surname||', '||m2.firstname) AS recfullname
FROM Members AS m1
LEFT JOIN Members as m2
ON m1.recommendedby = m2.memid
WHERE m1.memid !=0
ORDER BY fullname
"""
displayQuery(db,q11)
```

```python
q12="""
SELECT f.name, mem_uses.memid, mem_uses.ct AS member_usage
FROM Facilities AS f
LEFT JOIN
(SELECT facid, memid, COUNT(*) AS ct
FROM Bookings 
WHERE memid != 0 
GROUP BY facid, memid) AS mem_uses
ON mem_uses.facid = f.facid
ORDER BY memid
"""
displayQuery(db,q12)
```

```python
#q12, but just to total usage by members per facility
q12alt="""
SELECT f.name, mem_uses.memid, mem_uses.ct AS member_usage
FROM Facilities AS f
LEFT JOIN
(SELECT facid, memid, COUNT(*) AS ct
FROM Bookings 
WHERE memid != 0 
GROUP BY facid) AS mem_uses
ON mem_uses.facid = f.facid
ORDER BY f.facid
"""
displayQuery(db,q12alt)
```

```python
q13="""
SELECT f.name, mem_uses.month, mem_uses.ct AS member_usage
FROM Facilities AS f
LEFT JOIN
(SELECT facid, CAST(STRFTIME('%m', starttime) AS INTEGER) AS month, COUNT(*) AS ct
FROM Bookings 
WHERE memid != 0 
GROUP BY facid, month) AS mem_uses
ON mem_uses.facid = f.facid
ORDER BY month
"""
displayQuery(db,q13)
```
