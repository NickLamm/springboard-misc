```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rawData = pd.read_csv('annual-number-of-deaths-by-cause.csv')
rawData[rawData.Code == 'OWID_USS']
```

```python
rawData = rawData[rawData.Code.notnull()].drop(columns = 'Code') #eliminate regional aggregations, then drop now superfluous col
rawData
```

```python
#This block runs very slowly because I wrote it in 3 minutes
#My options were to let it run and go for a walk or spend the equal time making it more efficient
data = pd.DataFrame(columns = ['Entity', 'Year', 'Cause', 'Deaths'])
i=0
for row in rawData.iterrows():
    break # delete to reprocess
    identifiers = row[1][['Entity', 'Year']]
    values = row[1].drop(['Entity', 'Year'])
    for cause in values.index:
        newRow = identifiers
        newRow['Cause'] = cause
        newRow['Deaths'] = values[cause]
        print(newRow)
        data.loc[len(data.index)] = newRow
#data.to_csv('processed data.csv', index=False)
data = pd.read_csv('processed data.csv')
data.Deaths = pd.to_numeric(data.Deaths.str.extract('(\d+)', expand=False))
#we accidentally left in entries for USSR and 'World', those can be removed
data = data[data.Entity != 'USSR']
data = data[data.Entity != 'World']
#also a surprising number of rows are missing in general
data.dropna(inplace=True)
#and also drop any countries who have less than 10000 listed deaths
#if a country is listed as having less than 10000 deaths over 12 years, either the data is bad
#or the country is too small to use for our purposes
ununsedEntity = (data.groupby('Entity').Deaths.sum() < 10000)
entitiesToDrop = ununsedEntity.index[ununsedEntity]
data = data[~data.Entity.isin(entitiesToDrop)]
data
```

```python
#this will graph the average deaths over time for each cause
avgByYear = pd.DataFrame(data.groupby(['Cause', 'Year']).Deaths.mean()).unstack().T.reset_index(level=0, drop=True)
_=avgByYear.plot(subplots = True, figsize = (14,100), sharex=False)
```

Many of these causes have changed steadily. The most notable exceptions are ones related to conflict, terrorism, and environmental exposure. Sudden changes can often be linked to historical tradgedies. For example, "Conflict and terrorism" has a large spike in the year 1994. This was the year of the Rwanadan genocide.

```python
ninetyfour = data[data.Year == 1994]
ninetyfour[ninetyfour.Cause == 'Conflict and terrorism'].sort_values('Deaths', ascending=False).head()
```

For the remaining causes, some increase is to be expected, since the population of the world also increases over time. Despite this, several causes display a constant downward trend. These primarily include deaths caused by infection or improper nutrition:

Acute hepetitis

Diarrheal diseases

Respiratory infections

Maternal Disorders

Meningitis

Neonatal disorders

Malnutrition of any kind

Tuberculosis

Prevention of all these things fall under the field of medicine, which has been steadily improving throughout history, so this is to be expected. However, drowning also has trended downwards over the entire oberseved time.

```python
drowning = data[data.Cause == 'Drowning'].reset_index(drop=True)
drowning.groupby('Entity').sum().Deaths.sort_values(ascending=False).head()
```

```python
_=drowning[drowning.Entity=='China'].sort_values('Year').set_index('Year').Deaths.plot(title='Drowning deaths: China')
```

```python
_=drowning[drowning.Entity=='India'].sort_values('Year').set_index('Year').Deaths.plot(title='Drowning deaths: India')
```

Because this downward trend can be attributed to the two countries with the highest drowning rates, it can be safely assumed that improvements to safety standards in those countries/the world are having a measurable effect.

Of the upward trending causes, not all trend at the same rate. For example, the rate of cardiovascular disease has risen much faster than chronic kidney disease.

```python
totalAvgChange = avgByYear.iloc[-1] - avgByYear.iloc[0]
_=totalAvgChange.plot(kind = 'bar', figsize=(12,8), title='Total change in average deaths')
```

This graph points to heart disease and neoplasms as the fastest rising killers in the world today.
