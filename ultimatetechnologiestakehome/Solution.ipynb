This makes sense because people are less likely to be willing or able to transport themselves when they are busy in the middle of the day or tired/inebriated late at night.
```python
with open('ultimate_data_challenge.json', 'r') as f:
    data = pd.DataFrame(json.load(f))
data
```

```python
data.isna().sum()
```

```python
data.nunique()
```

```python
#missing average ratings are probably people who just don't rate every time. Make them 0.
data.avg_rating_of_driver = data.avg_rating_of_driver.fillna(0)
data.avg_rating_by_driver = data.avg_rating_by_driver.fillna(0)
```

```python
#missing phone will be solved incidentally by changing to one-hot
data = data.drop(columns=['city','phone']).join(pd.get_dummies(data.loc[:,['city', 'phone']]))
data.isna().sum()
```

```python
data.dtypes
```

```python
#convert signup_date, last_trip_date, and ultimate_black_user cols
data.signup_date = pd.to_datetime(data.signup_date)
data.last_trip_date = pd.to_datetime(data.last_trip_date)
data.ultimate_black_user = data.ultimate_black_user.astype(int)
data.dtypes
```

```python
#we must now determine which users are active
#to do this, we will assume at least one user took a trip on the day the data was pulled
#that means any user with a last_trip_date within 30 days of the latest recorded last_trip_date is considered active
data['active'] = (data.last_trip_date.max() - data.last_trip_date <= timedelta(days=30)).astype(int)
data.drop(columns='last_trip_date', inplace=True)
```

```python
#in order for sklearn to use signup_date we will convert it to a simpler int column, signup_day
data['signup_day'] = (data.signup_date - datetime(2014,1,1)).dt.days
data.drop(columns='signup_date', inplace=True)
```

```python
data
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.pipeline import Pipeline
```

```python
X = data.drop(columns='active')
y = data.active
featSelector = Pipeline([
    ('scaler', MinMaxScaler()),
    ('selector', SelectKBest(score_func=chi2, k='all'))
])
featSelector.fit(X, y)
features = pd.DataFrame({'Feature':featSelector.named_steps.scaler.feature_names_in_,'P-value':featSelector.named_steps.selector.pvalues_})
features.sort_values('P-value', ascending=True).reset_index(drop=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=191)
rfpipe = Pipeline([
    ('scaler', MinMaxScaler()),
    ('selector', SelectKBest(score_func=chi2)),
    ('random_forest', RandomForestClassifier())
])
gbpipe = Pipeline([
    ('scaler', MinMaxScaler()),
    ('selector', SelectKBest(score_func=chi2)),
    ('random_forest', GradientBoostingClassifier())
])
kbestparams = {'selector__k': [4,5,6,7,8,'all']}
rfsearch = GridSearchCV(estimator=rfpipe, param_grid=kbestparams)
gbsearch = GridSearchCV(estimator=gbpipe, param_grid=kbestparams)
```

```python
rfsearch.fit(X_train, y_train)
```

```python
gbsearch.fit(X_train, y_train)
```

```python
gradientBoostModel = gbsearch.best_estimator_
print(gbsearch.best_params_)
gradientBoostModel.score(X_test, y_test)
```

```python
randomForestModel = rfsearch.best_estimator_
print(rfsearch.best_params_)
randomForestModel.score(X_test, y_test)
```

Gradient boosting model provides slightly better performance concerning accuracy.
